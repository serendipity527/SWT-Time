```python
# 测试脚本
import torch
import time
import ptwt

x = torch.randn(32, 7, 512).cuda()

# SWT 版本
start = time.time()
swt_coeffs = ptwt.swt(x, 'db4', level=3)
swt_features = [c.mean() for c in swt_coeffs]
swt_time = time.time() - start

# DWT 版本
start = time.time()
dwt_coeffs = ptwt.wavedec(x, 'db4', level=3)
dwt_features = [c.mean() for c in dwt_coeffs]
dwt_time = time.time() - start

print(f"SWT: {swt_time*1000:.2f}ms, DWT: {dwt_time*1000:.2f}ms")
print(f"特征差异: {[abs(s-d).item() for s,d in zip(swt_features, dwt_features)]}")
```
